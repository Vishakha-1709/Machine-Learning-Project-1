```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df.shape
```

```python
df.drop(['class', 'who', 'adult_male', 'deck', 'embark_town','alive'],inplace=True,axis=1)
```

```python
df.info()
```

```python
df['age']=df['age'].fillna(df['age'].mean())
```

```python
#drop columns for embarked
df.dropna(subset=['embarked'],inplace=True)
```

```python
df.shape
```

```python
df.head()
```

```python
df['sibsp'].value_counts()
```

```python
#label encoding
# df['sex']=df['sex'].map({"male":0,"female":1})

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()

df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
df['embarked'].value_counts()
```

```python
df.head()
```

```python
df=df.astype(int)
df
```

```python
x=df.drop("survived",axis=1)
y=df['survived']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred=y_pred.astype(int)
y_pred
```

logical Regression


```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
accuracy_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

KNN MODEL

```python
# to use this we should do  feature extraction 
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train_scaled=scaler.fit_transform(x_train)
x_test_scaled=scaler.fit_transform(x_test)
x_test_scaled
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn_model=KNeighborsClassifier(n_neighbors=5)
knn_model.fit(x_train_scaled,y_train)
```

```python
y_knn = knn_model.predict(x_test_scaled)
```

```python
accuracy_score(y_test,y_knn)
```

```python
confusion_matrix(y_test,y_knn)
```

```python
print(classification_report(y_test,y_knn))
```
